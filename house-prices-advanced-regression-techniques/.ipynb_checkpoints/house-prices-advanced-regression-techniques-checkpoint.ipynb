# Analyze missing values

```python
null_values = train.isnull().sum()
null_values[null_values > 0]
```

```python
fig = plt.figure(figsize=(17, 17))
sns.heatmap(train.isnull(), cbar=False, yticklabels=False, cmap='terrain')
```
